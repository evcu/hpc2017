# Read results and report

```python
from collections import defaultdict
d=defaultdict(list)
with open('results.out','rb') as f:
    print f.next()
    for l in f:
        nthread,N = map(int,l.split())
        time = float(f.next().split()[3])
        d[nthread].append((N,time))
print d
```

```python
import matplotlib.pyplot as plt

for k,v in d.items():
    x = [vv[0] for vv in v]
    y = [vv[1] for vv in v]
    plt.plot(x, y,label=k)
plt.legend(loc=0)
plt.xlabel("Time in seconds")
plt.ylabel("N")
plt.title("Time spent for 5000 iteration for different #threads and N for Jacobi smoothing")
plt.show()
```
